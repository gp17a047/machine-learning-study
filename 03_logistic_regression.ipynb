## ロジスティック回帰とは
　今回は、前回、前々回の単回帰分析、重回帰分析を読んだことを前提の内容となっております。ご了承ください。

　ロジスティック回帰は、主に確率の問題や分類問題を扱いたいときに使われます。確率や分類は 0か1、0%〜100%といった一定の範囲で扱われます。例えば降水確率は 0〜100%ですし、分類の場合は、この写真が猫であるか否か、この写真が犬であるか否かという0, 1で表されます（0.0~1.0の確率で表すこともあります）。
 
 その場合、今まで学んだ単回帰分析、重回帰分析では扱いにくいため、ロジスティック回帰というものが用いられます。今までの予測に加えて、分類にも使いやすいのがポイントですね（回帰という名前が紛らわしいですが）。
 
 

## シグモイド関数
　ロジスティック回帰では、結果を一定の範囲（具体的には0〜1）に制限するために、シグモイド関数と呼ばれる関数を使います。シグモイド関数は、ロジスティック関数とも呼ばれます。同じ関数なのに、違う名前がついているのは、歴史的経緯（使われている分野が異なるため）です。興味がある方は調べてみて下さい。
 
 　シグモイド関数の式は以下です。

```math
f(x) = \frac{1}{1 + e^{-ax}}
```

````python
import numpy as np
import matplotlib.pylab as plt
import seaborn as sns

def sigmoid(x):
    return 1/(1 + np.exp(-x))

x = np.arange(-5.0, 5.0 , 0.1)
y = sigmoid(x)
plt.plot(x, y)
plt.ylim(-0.1, 1.1)
plt.show()
````
